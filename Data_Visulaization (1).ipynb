```python
#Line Plot



import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline



#define dataset

x= np.arange(1,10,0.1)
y= 7*x+1


#plot the data Points

plt.plot(x,y)
plt.show()
```

```python
#Bar plot



import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


#define dataset

data= {'apples':20 , 'orange': 10 , 'sapota': 25}
names=list(data.keys())
values=list(data.values())
fig = plt.figure(figsize=(6,5))   #The Figure size can be mentioned in this Form



plt.bar(names,values)
plt.show()




```

```python
#Bar plot    #Doubt in this   # Analysing matters



import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


#define dataset

data= {'banana': 20, 'Tomato': 15, 'Lemon': 30, 'litchi' : 10 }
names=list(data.keys())
values=list(data.values())




fig = plt.figure(figsize=(10,5))   #The Figure size can be mentioned in this Form






plt.barh(names,values,color="yellow")
plt.title('Horizantal Bar Plot')
plt.xlabel("Quantity")
plt.ylabel("Fruits")
plt.show()
```

```python
#Histogram



import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


numbers=[2,15,13,14,19,110,16,17,18,40,45,25,90,65,65,56,75,13,49,13,4,71,36,46,58,15,35,64,97,21]



plt.hist(numbers,bins=[0,20,40,60,80,100], edgecolor='#FFFFFF', color='#FF2331')

plt.title('Histogram Plot')
plt.xlabel("Numbers")
plt.ylabel("Bins")
plt.grid(True)
plt.show()

```

```python
#Box Plot


import matplotlib.pyplot as plt
%matplotlib inline


#Data Prep
Total=[11,28,32,65,56,36,98,20,58,38,41,37,7,39]
Order=[70,41,53,90,78,3,62,50,35,60,54,3,1]
Dist=[8,23,69,28,87,13,5,97,31,58,84,4,50]
data= list ([Total,Order,Dist])       #Organise in List



#Plotting the data


plt.boxplot(data, showmeans= True)
plt.title('Box plot')
plt.grid(True)
plt.show



```

```python
import matplotlib.pyplot as plt
import statistics



%matplotlib inline
#Data Prep
Total=[11,28,32,65,56,36,98,20,58,38,41,37,7,39]

x = statistics.median(Total)
print ("Mean",x)

plt.boxplot(Total, showmeans= True)

plt.title('BOx plot')
plt.grid(True)
plt.show




```

```python
#Violin Plot


import matplotlib.pyplot as plt
%matplotlib inline


#Data Prep
Total=[11,28,32,65,56,36,98,20,58,38,41,37,7,39]
Order=[70,41,53,90,78,3,62,50,35,60,54,3,1]
Dist=[8,23,69,28,87,13,5,97,31,58,84,4,50]
data= list ([Total,Order,Dist])       #Organise in List



#Plotting the data


plt.violinplot(data, showmeans= True , showmedians= True)
#this shows showmeans= True , showmedians= True
#Mean and Median
plt.title('Violin plot')
plt.grid(True)
plt.show
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Intialize the Data

Label = ['Raja','Rani','Police','Thirudan']
Size = [10000,5000,1000,500]


#Plotting the data
#To display the Percentage "autopct='%1.1f%%"
#To add Shadows "shadow=True"

plt.pie(Size,labels=Label,autopct='%1.1f%%',shadow=True)
plt.title('Pie plot')
plt.show()




```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Intialize the Data

Label = ['Raja','Rani','Police','Thirudan']
Size = [10000,5000,1000,500]
Size_center= [15]


#Plotting the data
#To display the Percentage "autopct='%1.1f%%"
#To add Shadows "shadow=True"

plt.pie(Size,labels=Label,autopct='%1.1f%%',shadow=True)
plt.pie(Size_center ,radius= 0.3, colors='white')
plt.title('Pie plot')
plt.show()
```
