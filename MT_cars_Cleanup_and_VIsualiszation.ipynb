```python
import numpy as np
import pandas as pd
path ="/content/mtcars.csv"
cars = pd.read_csv(path)
cars
```

```python
#rename column
cars=cars.rename (columns={'Unnamed: 0': 'model'})
cars
```

```python
#fill in null values with mean of the column
cars.qsec=cars.qsec.fillna(cars.qsec.mean())
cars
```
